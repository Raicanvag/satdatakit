```python
# [1] Setup
from satdatakit import read, compute_index
from satdatakit.extensions.dask_ext import enable_dask, read_dask
import warnings
warnings.filterwarnings("ignore", category=UserWarning)

enable_dask()
print("✅ Dask ready")
```

```python
# [2] read_dask directo
ds = read_dask(
    ["/workspace/data/sentinel2_test.tif"],
    chunks={"x": 256, "y": 256}
)
print(f"Type: {type(ds.data.data)}")
print(f"Shape: {ds.shape}")
print(f"Bands: {ds.bands}")
```

```python
# [3] to_dask() en dataset existente
ds2 = read("/workspace/data/sentinel2_test.tif")
print(f"Before: {type(ds2.data.data)}")
ds2 = ds2.to_dask(chunks={"x": 256, "y": 256})
print(f"After: {type(ds2.data.data)}")
```

```python
# [4] Índice lazy
ds_ndvi = compute_index(ds2, "NDVI")
print(f"Lazy NDVI: {type(ds_ndvi.data.data)}")
```

```python
# [5] Computar
ds_ndvi = ds_ndvi.compute()
print(f"Computed: {type(ds_ndvi.data.data)}")
```

```python
# [6] Tabla resumen
print("""
| Escenario | Usar |
|---|---|
| 1 archivo < 1 GB | Core (sin Dask) |
| 10-100 archivos | read_dask() + compute() |
| Archivo > 10 GB | to_dask(chunks={"x": 1024}) |
| Cluster multi-nodo | dask.distributed.Client() |
""")
```
